```python
import os
import sys
import json
import time
import cv2
import torch
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

from skimage.measure import compare_ssim as ssim

sys.path.append('..')
from utils.model import CSRNet
from utils.augmentation import create_dataloader
```

```python
%matplotlib inline
```

```python
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'
print('Using {} as device'.format(device))
```

```python
# Load trained model
CKPT_DIR = '../ckpts'
CKPT_FILENAME = 'model_A-50.18.pth.tar'
CKPT_PATH = os.path.join(CKPT_DIR, CKPT_FILENAME)

ckpt = torch.load(CKPT_PATH, map_location=torch.device(device))
model = CSRNet(training=False)
model.load_state_dict(ckpt['state_dict'])
```

```python
# Generate dataset
DATA_DIR = '../part_B'
DATA_FILENAME = 'test.json'
DATA_PATH = os.path.join(DATA_DIR, DATA_FILENAME)

with open(DATA_PATH) as infile:
    image_paths = json.load(infile)
    image_paths = [os.path.join('..', path) for path in image_paths]

dataloader = create_dataloader(image_paths, 'validation', shuffle=False, batch_size=1)
dataset = iter(dataloader)
idx = 0
```

```python
def show_batch(idx, n_images=3):
    fig, ax = plt.subplots(nrows=3, ncols=n_images, figsize=(10, 10), gridspec_kw={'wspace': 0, 'hspace': 0})
    
    for i in range(n_images):
        # Fetch data
        raw_image = cv2.imread(image_paths[i + idx])
        image, target = next(dataset)
        start_time = time.time()
        output = model(image)
        end_time = time.time()
        print('Density estimation took {:0.2f} secs'.format(end_time - start_time))
        
        # Format correctly to matplotlib
        raw_image = cv2.cvtColor(raw_image, cv2.COLOR_BGR2RGB)
        target = target.squeeze().numpy()
        output = output.squeeze().detach().numpy()
        
        # Fetch data
        gt_counts = target.sum()
        estimated_counts = output.sum()
        mae = abs(gt_counts - estimated_counts)
        instance_ssim = ssim(target, output)
        
        # Log info
        image_info = 'Image:{}'.format(i)
        gt_info = ':Gt:{:0.2f}'.format(gt_counts)
        est_info = ':Output:{:0.2f}'.format(estimated_counts)
        mae_info = ':MAE:{:0.2f}'.format(mae)
        print(image_info + gt_info + est_info + mae_info)
        
        # Resize images
        raw_image = cv2.resize(raw_image, (512, 512))
        target = cv2.resize(target, (512, 512))
        output = cv2.resize(output, (512, 512))
        
        # Show images
        ax[0, i].imshow(raw_image)
        ax[0, i].axis('off')
        ax[0, i].set_xticklabels([])
        ax[0, i].set_yticklabels([])
        ax[0, i].set_aspect('equal')
        
        ax[1, i].imshow(target, cmap=cm.jet)
        ax[1, i].axis('off')
        ax[1, i].set_xticklabels([])
        ax[1, i].set_yticklabels([])
        ax[1, i].set_aspect('equal')
        
        ax[2, i].imshow(output, cmap=cm.jet)
        ax[2, i].axis('off')
        ax[2, i].set_xticklabels([])
        ax[2, i].set_yticklabels([])
        ax[2, i].set_aspect('equal')
        
        # Annotate results
        ax[2, i].annotate('Count: {:0.0f}'.format(gt_counts), (0, 0), (0, -5),
                          xycoords='axes fraction', textcoords='offset points', 
                          va='top')#, fontweight='semibold')
        ax[2, i].annotate('Estimate: {:0.2f}'.format(estimated_counts), (0, 0), (0, -17),
                          xycoords='axes fraction', textcoords='offset points', 
                          va='top')#, fontweight='semibold')
        ax[2, i].annotate('SSIM: {:0.2f}'.format(instance_ssim), (0, 0), (0, -29),
                          xycoords='axes fraction', textcoords='offset points', 
                          va='top')#, fontweight='semibold')
    
    # plt.tight_layout()
    plt.show()
```

```python
n_images = 3
```

```python
show_batch(idx, n_images)
idx += n_images
```

```python
def show_gt_batch(idx, n_images=3):
    fig, ax = plt.subplots(ncols=n_images, nrows=2, figsize=(10, 10), gridspec_kw={'wspace': 0, 'hspace': 0})
    
    for i in range(n_images):
        raw_image = cv2.imread(image_paths[i + idx])
        image, target = next(dataset)
        
        # Format target to matplotlib
        target = target.squeeze().numpy()
        raw_image = cv2.cvtColor(raw_image, cv2.COLOR_BGR2RGB)
        
        # Resize
        raw_image = cv2.resize(raw_image, (512, 512))
        target = cv2.resize(target, (512, 512))
        
        # Show images
        ax[0, i].imshow(raw_image)
        ax[0, i].axis('off')
        ax[0, i].set_xticklabels([])
        ax[0, i].set_yticklabels([])
        ax[0, i].set_aspect('equal')
        
        ax[1, i].imshow(target, cmap=cm.jet)
        ax[1, i].axis('off')
        ax[1, i].set_xticklabels([])
        ax[1, i].set_yticklabels([])
        ax[1, i].set_aspect('equal')
    
    plt.show()
```

```python
show_gt_batch(idx, n_images=n_images)
idx += n_images
```
